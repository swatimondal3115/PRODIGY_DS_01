```python

# Filter out rows with NA values in '2020' and '1960'
clean_data = data.dropna(subset=['2020', '1960'])

# Sort the data by population in 2020 and select the top 10 countries
top_10_countries = clean_data.nlargest(10, '2020')

# Scatter plot for population growth from 1960 to 2020
top_10_countries['Population Growth'] = top_10_countries['2020'] - top_10_countries['1960']

plt.figure(figsize=(14, 10))
sns.scatterplot(x=top_10_countries['Country Name'], y=top_10_countries['Population Growth'])
plt.title('Population Growth from 1960 to 2020')
plt.xlabel('Country')
plt.ylabel('Population Growth')
plt.xticks(rotation=90)
plt.show()

# Histogram of population in 2020
plt.figure(figsize=(14, 10))
sns.histplot(clean_data['2020'], bins=30, kde=False, color='blue')
plt.title('Histogram of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()

# Define East Asia & Pacific countries
east_asia_pacific_countries = ["China", "Japan", "Korea, Rep.", "Indonesia", "Viet Nam"]

# Filter data for these countries
east_asia_pacific_data = clean_data[clean_data['Country Name'].isin(east_asia_pacific_countries)]

# Sort the data by population in 2020 and select the top countries
top_10_east_asia_pacific = east_asia_pacific_data.nlargest(10, '2020')

# Bar plot of top countries in East Asia & Pacific by population in 2020
plt.figure(figsize=(14, 10))
sns.barplot(x=top_10_east_asia_pacific['Country Name'], y=top_10_east_asia_pacific['2020'], palette='Oranges')
plt.title('Top Countries in East Asia & Pacific by Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

```python
# Filter out rows with NA values in '2023'
clean_data = data.dropna(subset=['2023'])

# Filter data for India and select columns from 2020 onwards
india_data = clean_data[clean_data['Country Name'] == "India"].loc[:, '2020':'2023']

# Reshape data to long format
india_data_long = india_data.melt(var_name='Year', value_name='Population')
india_data_long['Year'] = india_data_long['Year'].astype(int)

# Plot time series for India
plt.figure(figsize=(12, 6))
sns.lineplot(data=india_data_long, x='Year', y='Population', marker='o', color='blue')
plt.title('Population Time Series of India')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Filter and arrange data for top 15 countries by population in 2023
top_15_countries_2023 = clean_data.nlargest(15, '2023')

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_15_countries_2023['2023'], labels=top_15_countries_2023['Country Name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 15 Countries by Population in 2023')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import numpy as np
```

```python
# Drop rows with missing values in specific columns of interest
merged_data_clean = merged_data.dropna(subset=['Region', 'IncomeGroup'])

# Impute missing values in numeric columns with mean or median
numeric_columns = merged_data_clean.select_dtypes(include=np.number).columns
merged_data_clean[numeric_columns] = merged_data_clean[numeric_columns].fillna(merged_data_clean[numeric_columns].median())
```

```python
# Count plot of regions
plt.figure(figsize=(12, 8))
sns.countplot(y='Region', data=merged_data_clean, order=merged_data_clean['Region'].value_counts().index)
plt.title('Count of Countries by Region')
plt.xlabel('Number of Countries')
plt.ylabel('Region')
plt.tight_layout()
plt.show()

# Count plot of income groups
plt.figure(figsize=(10, 6))
sns.countplot(y='IncomeGroup', data=merged_data_clean, order=merged_data_clean['IncomeGroup'].value_counts().index)
plt.title('Count of Countries by Income Group')
plt.xlabel('Number of Countries')
plt.ylabel('Income Group')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'merged_data_clean' is your cleaned and processed DataFrame
countries_of_interest = ['India', 'China', 'United States', 'Brazil', 'Nigeria']

plt.figure(figsize=(12, 8))

for country in countries_of_interest:
    country_data = merged_data_clean[merged_data_clean['Country Name'] == country]
    years = country_data.columns[4:-4]
    population_values = country_data.iloc[:, 4:-4].values.flatten().astype(float)
    plt.plot(years, population_values, marker='o', label=country)

plt.title('Population Trends Over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
